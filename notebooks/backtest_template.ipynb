# 配置驱动的单标的回测模板

本 Notebook 只负责填写配置和展示结果。数据读取与自动补数、策略构建、回测、基准比较、绩效计算和绘图均由 `quant_backtest` 公共工具完成。

修改配置单元后依次运行即可。美股基准可使用 `SPY.US`，A 股基准可使用沪深300 ETF `510300.SH`。

```python
from quant_backtest import (
    BacktestReportConfig,
    BacktestReportPlotter,
    MovingAverageResearchConfig,
    SingleAssetResearchRunner,
)
```

```python
config = MovingAverageResearchConfig(
    code="NVDA.US",
    benchmark_code="SPY.US",
    start="2018-01-01",
    end="2024-12-31",
    short_window=5,
    long_window=20,
    transaction_cost_bps=5.0,
    risk_free_rate=0.0,
    auto_download=True,
    report=BacktestReportConfig(
        panels=("signals", "equity", "drawdown"),
        title="NVDA Moving Average Strategy",
        save_path=None,
    ),
)
```

```python
runner = SingleAssetResearchRunner()
result = runner.run(config)
```

```python
report = BacktestReportPlotter().build(result)
display(report.metrics)
display(report.relative_metrics.to_frame())
```
