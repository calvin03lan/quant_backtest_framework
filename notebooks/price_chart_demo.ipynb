# 标的价格图表

本 Notebook 从本地 MongoDB 的 `quant_db.daily_ohlc` 读取单一标的行情，并使用 `PriceChartEngine` 绘制静态折线图或 K 线图。

使用前请在项目根目录安装依赖并确认 MongoDB 已启动：

```bash
conda run -n quant pip install -e ".[notebook]"
```

K 线使用自适应频率：不超过 160 根为日 K，161–750 根为周 K，更长区间为月 K。图表标题会显示实际频率。

````
from quant_backtest import PriceChartEngine
from quant_backtest.data import MongoMarketDataRepository

repository = MongoMarketDataRepository()
repository.ping()
engine = PriceChartEngine(repository)

available_codes = engine.available_codes()
print(f"数据库中共有 {len(available_codes)} 个有行情的标的")
print(available_codes[:30])
````

## 参数

修改下面的参数后重新运行绘图单元：

- `CHART_TYPE`：`"line"` 或 `"candlestick"`
- `PRICE_COLUMN`：折线图可选 `open/high/low/close`，默认 `close`
- `ADJUSTED`：是否使用复权价格
- `FREQUENCY`：K 线可选 `auto/daily/weekly/monthly`
- `SHOW_VOLUME`：K 线是否显示成交量
- `SAVE_PATH`：设为 PNG 路径时保存图片，`None` 表示只显示

````
# === 用户参数 ===
CODE = "AAPL.US"
START_DATE = "2023-01-01"
END_DATE = "2024-12-31"
CHART_TYPE = "line"  # "line" 或 "candlestick"
PRICE_COLUMN = "close"  # 折线图默认收盘价
ADJUSTED = False
FREQUENCY = "auto"  # "auto" / "daily" / "weekly" / "monthly"
SHOW_VOLUME = True
SAVE_PATH = None  # 例如 "outputs/aapl_chart.png"
````

````
import matplotlib.pyplot as plt

figure, axes = engine.plot(
    CODE,
    START_DATE,
    END_DATE,
    chart_type=CHART_TYPE,
    price_column=PRICE_COLUMN,
    adjusted=ADJUSTED,
    frequency=FREQUENCY,
    show_volume=SHOW_VOLUME,
    save_path=SAVE_PATH,
)

if SAVE_PATH:
    print(f"图片已保存到：{SAVE_PATH}")
plt.show()
````
